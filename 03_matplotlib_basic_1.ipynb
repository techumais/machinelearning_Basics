```python
mu,sigma = 100,15
data = mu+sigma*np.random.random(10000)
plt.hist(data,50,normed=1,facecolor='g',alpha=0.75)
plt.plot([10,20,30,40,50],[100,200,300,400,500],".g")
plt.title("Histogram")
plt.xlabel("Smart")
plt.ylabel("Probability")
plt.grid()
plt.show()
```
